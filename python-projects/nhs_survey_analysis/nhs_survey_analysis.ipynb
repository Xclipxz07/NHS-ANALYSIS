```python
# nhs_survey_analysis.py
# Example: simple analysis of mock NHS patient satisfaction survey
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'respondent_id': range(1,11),
    'overall_satisfaction': [4,5,3,4,2,5,4,3,4,5],
    'waiting_time_minutes': [30,45,120,60,90,20,25,60,40,15],
    'would_recommend': ['Yes','Yes','No','Yes','No','Yes','Yes','No','Yes','Yes']
}
df = pd.DataFrame(data)

print('Survey head:')
print(df.head())
```

```python

# Average satisfaction
print('\nAverage satisfaction:', df['overall_satisfaction'].mean())
```

```python

# Correlation between waiting time and satisfaction
corr = df['overall_satisfaction'].corr(df['waiting_time_minutes'])
print('\nCorrelation (satisfaction vs waiting time):', corr)
```

```python

# Simple plot saved to file
plt.figure(figsize=(6,4))
plt.scatter(df['waiting_time_minutes'], df['overall_satisfaction'])
plt.xlabel('Waiting time (minutes)')
plt.ylabel('Satisfaction (1-5)')
plt.title('Waiting time vs Satisfaction (Mock Data)')
plt.tight_layout()
plt.savefig('waiting_vs_satisfaction.png')
print('\nPlot saved to waiting_vs_satisfaction.png')
```
