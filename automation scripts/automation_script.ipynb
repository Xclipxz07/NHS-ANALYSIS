# Automation Script
Examples of automating a small report generation and KPI summary from tasks CSV.

```python
import pandas as pd
df = pd.read_csv('automation_tasks.csv', parse_dates=['date'])
print('Loaded', len(df), 'tasks')
display(df.head())
```

```python
# Generate status counts
status_counts = df['status'].value_counts().reset_index()
status_counts.columns = ['status','count']
status_counts.to_csv('automation_status_counts.csv', index=False)
print('Saved automation_status_counts.csv')
display(status_counts)
```

```python
# Simple weekly report: tasks per week
df['week_start'] = df['date'].dt.to_period('W').apply(lambda r: r.start_time)
weekly = df.groupby('week_start').size().reset_index(name='tasks')
weekly.to_csv('automation_weekly_report.csv', index=False)
print('Saved automation_weekly_report.csv')
display(weekly)
```
